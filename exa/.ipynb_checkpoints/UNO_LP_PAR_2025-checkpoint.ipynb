1. **[2 puntos]** Aplíquese programación funcional para el siguiente problema: se tiene un archivo de datos con registros de correo electrónico. Cuéntese correos electrónicos provienen de cada dominio.

````python
from collections import Counter

correos = ['a@mecon.gob.ar', 'b@mecon.gob.ar', 'c@afip.gob.ar', 'd@mecon.gob.ar']

# Extraer dominios usando map
dominios = list(map(lambda correo: correo.split('@')[-1], correos))

# Contar ocurrencias con Counter
conteo = Counter(dominios)

# Mostrar resultados
for dominio, cantidad in conteo.items():
    print(f"{dominio}: {cantidad}")
````

2. **[2 puntos]** Aplíquese programación lógica para el siguiente problema: aprobar un parcial, teniendo en cuenta que lo puede aprobar un alumno que lee el material, asiste a clases, hace ejercicios. También lo puede aprobar un sabelotodo. También quien tenga un tutor que haya aprobado el parcial. Alex y Maria aprueban el parcial.

````python
import janus_swi as janus
janus.consult('parcial', '''
aprueba_parcial(X) :- alum(X), lee_material(X), asiste_a_clases(X), realiza_ejercicios(X).
aprueba_parcial(X) :- alum(X), sabe_todo(X).
aprueba_parcial(X) :- alum(X), tutor(Y,X), aprueba_parcial(Y), !.
alum(maria).
alum(alfredo).
alum(tomas).
alum(susana).
alum(juan).
alum(alex).
lee_material(alfredo).
lee_material(maria).
asiste_a_clases(alfredo).
asiste_a_clases(maria).
realiza_ejercicios(susana).
realiza_ejercicios(maria).
sabe_todo(alex).
tutor(maria,juan).
tutor(juan,susana).
''')
list(janus.query('aprueba_parcial(X).'))
````
